1.ScatterPlot(Numerical-Numerical)

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue=df["sex"],data=df)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue=df["sex"],style=df["smoker"],data=df)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue=df["sex"],style=df["smoker"],size=df["size"],data=df)
```

2.BarPlot(Numerical-Numerical)

```python
sns.barplot(x='Pclass',y='Age',data=titanic)
```

```python
sns.barplot(x='Pclass',y='Age',hue=titanic["Sex"],data=titanic)
```

3.BoxPlot(Numerical-Categorical)

```python
sns.boxplot(x='Sex',y='Age',data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',data=titanic,hue=titanic["Survived"])
```

4.DistPlot(Numerical-Categorical)

```python
sns.distplot(titanic["Age"])
```

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"])
sns.distplot(titanic[titanic["Survived"]==1]["Age"])
```

```python
sns.distplot(titanic[titanic["Survived"]==1]["Age"])
```

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"],hist=False)
sns.distplot(titanic[titanic["Survived"]==1]["Age"],hist=False)
```

5.HeatMap(CAtegorical-Categorical)

```python
pd.crosstab(titanic["Pclass"],titanic["Survived"])
```

```python
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
```

```python
titanic.groupby("Pclass")["Survived"].mean()*100
```

```python
(titanic.groupby("Pclass")["Survived"].mean()*100).plot(kind="bar")
```

6.ClusterMap(Categorical-Categorical)

```python
pd.crosstab(titanic["SibSp"],titanic["Survived"])
```

```python
sns.clustermap(pd.crosstab(titanic["SibSp"],titanic["Survived"]))
```

7.PairPlot

```python
sns.pairplot(iris,hue="species")
```
